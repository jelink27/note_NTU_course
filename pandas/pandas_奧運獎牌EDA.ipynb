```python
# Q1 哪個國家贏得夏季奧運金牌數最多

df[df['Gold'] == df['Gold'].max()]


def answer_one(df):
  return df[df['Gold'] == df['Gold'].max()].index

answer_one(df)  
```

```python
# Q2 哪個國家夏季奧運和冬季奧運金牌數差距最大

#max的Gold-Gold.2

def answer_two(df):
  df['gold_diff'] = abs(df['Gold.1'] - df['Gold'])
  return df['gold_diff'].idxmax()

answer_two(df)
```

```python
#Q3 哪個國家夏季奧運與冬季奧運的金牌數差距除以總金牌數的比例最大？ (僅考慮至少有一個夏季金牌與一個冬季金牌的國家)

df = df[df['Gold'] != 0]
df = df[df['Gold.1'] != 0]
df.shape

df['gold_ratio'] = abs(df['Gold.1'] - df['Gold']) / df['Gold.2']
df['gold_ratio'].idxmax()
df.loc['Bulgaria']
```

```python
#Q4 計算146個國家的獎牌點數，金牌3點，銀牌2點、銅牌1點

def answer_four(df):
  df['total_score'] = df['Gold.2']*3 + df['Silver.2']*2 + df['Bronze.2']*1
  return df
answer_four(df)
print('最高分的是',df['total_score'].max())
print('最高分的國家是',df['total_score'].idxmax())
print('最低分的是',df['total_score'].min())
print('最低分的國家是',df['total_score'].idxmin())
```
